# 01. Data Exploration

Solar Power Generation Data 분석

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화 설정
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. 데이터 로드

Kaggle에서 데이터 다운로드:
https://www.kaggle.com/datasets/anikannal/solar-power-generation-data

`data/` 폴더에 저장

```python
# 발전 데이터
generation = pd.read_csv('../data/Plant_1_Generation_Data.csv')
print(f"Generation data shape: {generation.shape}")
generation.head()
```

```python
# 날씨/센서 데이터
weather = pd.read_csv('../data/Plant_1_Weather_Sensor_Data.csv')
print(f"Weather data shape: {weather.shape}")
weather.head()
```

## 2. 기본 정보 확인

```python
print("=== Generation Data ===")
print(generation.info())
print("\n=== Weather Data ===")
print(weather.info())
```

```python
# 통계 요약
generation.describe()
```

## 3. 시각화

```python
# 시간 변환
generation['DATE_TIME'] = pd.to_datetime(generation['DATE_TIME'])
weather['DATE_TIME'] = pd.to_datetime(weather['DATE_TIME'])
```

```python
# 발전량 시계열
plt.figure(figsize=(14, 5))
plt.plot(generation['DATE_TIME'], generation['DC_POWER'])
plt.title('DC Power Generation Over Time')
plt.xlabel('Date')
plt.ylabel('DC Power (kW)')
plt.show()
```
